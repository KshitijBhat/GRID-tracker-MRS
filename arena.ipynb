```python
def get_arena_img():
    grid = np.zeros((20*6*2,10*6*2))
    grid[0:8*6*2,0:8*6*2] = 1
    grid[(8+4)*6*2:20*6*2,0:8*6*2] = 1
    for i in range(20):
        grid[i*6*2:i*6*2+1,0:10*6*2] = 1
    for i in range(10):
        grid[0:20*6*2,i*6*2:i*6*2+1] = 1    
        
    fig,ax = plt.subplots(figsize = (10,5))
    ax.imshow(grid.T,cmap = plt.cm.Dark2)
    plt.show()           
    
get_arena_img()    
```

```python
def get_arena():
    grid = np.zeros((20*6*2,10*6*2))
    grid[0:8*6*2,0:8*6*2] = 1
    grid[(8+4)*6*2:20*6*2,0:8*6*2] = 1
    for i in range(20):
        grid[i*6*2:i*6*2+1,0:10*6*2] = 1
    for i in range(10):
        grid[0:20*6*2,i*6*2:i*6*2+1] = 1   
    return grid    
```

```python
def get_grid():
    grid = np.zeros((20*6*2,10*6*2))
    grid[0:8*6*2,0:8*6*2] = 1
    grid[(8+4)*6*2:20*6*2,0:8*6*2] = 1   
    return grid
```

```python
import purepursuit
purepursuit.main()
```

```python
def lattice_to_coord(latt):
    x,y = latt
    return x*6*2+6,y*6*2+6
```

```python
fig,ax = plt.subplots(figsize = (10,5))
ax.imshow(get_arena().T,cmap = plt.cm.Dark2)
x,y = lattice_to_coord((8,0))
ax.scatter(x,y, marker = ".", color = "yellow", s = 200)
x1,y1 = lattice_to_coord((2,8))
ax.scatter(x1,y1, marker = ".", color = "red", s = 200)
plt.show()
```

```python
path = [("straight",8*6*2-6),("turn",90),("straight",7*6*2-6)]
```

```python
x,y = lattice_to_coord((8,0))
x1,y1 = lattice_to_coord((1,8))
```

```python
xp,yp,tp = generate_trajectory(path,(x,y,np.pi/2),radius = 0.5*6*2)
```

```python
fig,ax = plt.subplots(figsize = (10,5))
ax.imshow(get_arena().T,origin = 'lower',cmap = plt.cm.Dark2)
ax.plot(xp,yp)

ax.scatter(x,y, marker = ".", color = "yellow", s = 200)

ax.scatter(x1,y1, marker = ".", color = "red", s = 200)
plt.show()
```

```python
x,y = lattice_to_coord((8,0))
path = [("straight",8*6*2-6),("turn",90),("straight",6*6*2-6)]
xp,yp,tp = generate_trajectory(path,(x,y,np.pi/2),radius = 0.5*6*2)

pose = (x,y,np.pi/2)
tracker = PurePursuitTracker(xp,yp,2.2)
traj = []
while True:
    print("v = ",tracker.v," w = ",tracker.w)
    pose = simulate_unicycle(pose,tracker.v,tracker.w)

    if tracker.update(*pose):
        print("ARRIVED!!")    
        break
    traj.append([*pose, tracker.w, tracker.ref_idx])  
  
xs,ys,ts,ws,ids = zip(*traj)
plt.figure()
plt.plot(xp,yp,label='Reference')
x0,y0,t0 = pose
plt.quiver(x0,y0, np.cos(t0), np.sin(t0),scale=12)
plt.plot(xs,ys,label='Tracked')

plt.quiver(x0,y0, np.cos(t0), np.sin(t0),scale=12)
plt.title('Pure Pursuit trajectory')
plt.legend()
plt.grid()     
plt.show()
    
```
